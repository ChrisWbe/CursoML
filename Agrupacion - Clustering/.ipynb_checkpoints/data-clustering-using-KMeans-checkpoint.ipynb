## K-Means
Algoritmo populares de ML a la hora de hascer Clustering - Sistema de aprendizaje no supervisado

K-means calcula la distancia entre uno de los puntos (rojos) y todas las personas (azules)
<img src="kmeans.JPG"/>

Se agrupan los individuos en los puntos màs cercanos, se calcula el centro de cada grupo y se re-agrupan nuevamenten basado en estos centros, esto proceso se repite.
<img src="kmeansgrupos.JPG"/>

Se requiere un amplio conocimiento del dominio para poner una cantidad adecuada de grupos (puntos)<br>
Como los puntos se generan aleatoriamente puede suceder que se repita el proceso hasta obtener un buen resultado

<h1 style="color:purple">Clustering with scikut-learn</h1>

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
from itertools import cycle, islice
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
import datetime
%matplotlib inline
```

```python
#se crea un dataFrame desde el archivo csv (Datos meteorológicos por minuto)
data = pd.read_csv('meteo/minuto.csv')
```

## Data sampling
Demasiadas filas en el dataset, se bajará la cantidad a un 10%

```python
#Se elije una de cada 10 filas, para esto elijo cada fila que dividido por 10 tiene resto 0 (el id)
sample_df = data[(data['rowID']%10)==0]
sample_df.shape
```

## Estadisticas

```python
#para que la descripción luzca mejer se puede transponer el dataset
sample_df.describe().transpose()
```

```python
#las columnas ran accumulation y rain duration tienen valores en cero
sample_df[sample_df['rain_accumulation']==0].shape
```

```python
sample_df[sample_df['rain_duration']==0].shape
```

## Limpiando datos

```python
del sample_df['rain_accumulation']
del sample_df['rain_duration']
```

```python
rows_before = sample_df.shape[0]
#se borran filas que contengan valor nulo
sample_df = sample_df.dropna()
rows_after = sample_df.shape[0]
print("Numero des filas eliminadas -> ",(rows_before-rows_after))
```

```python
sample_df.shape
```

```python
sample_df.columns
```

```python
#Se selecciona los features o datos de interes o caracteristicas para encarar el clustering
features = ['air_pressure', 'air_temp',
       'avg_wind_direction', 'avg_wind_speed', 'max_wind_direction',
       'max_wind_speed','relative_humidity']
```

```python
#se filtra el contenido dejando solo las columnas de interes
select_df = sample_df[features]
```

```python
select_df.columns
```

```python
#Escalando los features
x = StandardScaler().fit_transform(select_df)
x
```

```python
#Implementando K-Means Clustering
#seteo la cantidad de cluster o puntos que se quieren generar
kmeans = KMeans(n_clusters=12)
#Ajuste del modelo
model = kmeans.fit(x) #Acá es donde ocurre el proceso, calcular centros, iteraar, reagrupar
print("model\n", model)
```

```python
#centros de los 12 grupos
centers = model.cluster_centers_
centers
```

## Plot
Se crean funciones para ayudar a trazar gráficos

```python
def pd_centers(featuresUsed, centers):
    colNames = list(featuresUsed)
    colNames.append('prediction')
    
    z = [np.append(A,index)for index,A in enumerate(centers)]
    p=pd.DataFrame(z,columns=colNames)
    p['prediction']=p['prediction'].astype(int)
    return p
```

```python
#Función para crear gráficos paralelos
def parallel_plot(data):
    my_colors = list(islice(cycle(['b','r','g','y','k']),None,len(data)))
    plt.style.use("seaborn-dark")
    plt.figure(figsize=(15,8)).gca().axes.set_ylim([-3,+3])
    parallel_coordinates(data,'prediction',color=my_colors,marker='o')
```

```python
p = pd_centers(features,centers) # se genera una tabla con los centros de cada columna
p
```

## Dry Days

```python
parallel_plot(p[p['relative_humidity'] < -0.5])
```

## Warm Days

```python
parallel_plot(p[p['air_temp'] > 0.5])
```

## Cool Days

```python
parallel_plot(p[(p['relative_humidity'] > 0.5) & (p['air_temp'] < 0.5)])
```
